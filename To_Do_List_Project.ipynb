```python
# To-Do List Application

# Initialize an empty to-do list
todo_list = []

# Function to display the current to-do list
def display_list():
    if not todo_list:
        print("Your to-do list is empty.")
    else:
        print("Your to-do list:")
        for i, item in enumerate(todo_list):
            print(f"{i + 1}. {item}")

# Function to add an item to the to-do list
def add_item():
    item = input("Enter the task to add: ")
    todo_list.append(item)
    print(f"Task '{item}' added to your to-do list.")

# Function to update an item in the to-do list
def update_item():
    display_list()
    if not todo_list:
        return
    index = int(input("Enter the index of the task to update: ")) - 1
    if index < 0 or index >= len(todo_list):
        print("Invalid index.")
    else:
        new_item = input("Enter the updated task: ")
        todo_list[index] = new_item
        print(f"Task '{new_item}' updated.")

# Function to delete an item from the to-do list
def delete_item():
    display_list()
    if not todo_list:
        return
    index = int(input("Enter the index of the task to delete: ")) - 1
    if index < 0 or index >= len(todo_list):
        print("Invalid index.")
    else:
        deleted_item = todo_list.pop(index)
        print(f"Task '{deleted_item}' deleted.")

# Main program loop
while True:
    print("\nMenu:")
    print("1. Display to-do list")
    print("2. Add task")
    print("3. Update task")
    print("4. Delete task")
    print("5. Quit")
    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        display_list()
    elif choice == "2":
        add_item()
    elif choice == "3":
        update_item()
    elif choice == "4":
        delete_item()
    elif choice == "5":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
```
